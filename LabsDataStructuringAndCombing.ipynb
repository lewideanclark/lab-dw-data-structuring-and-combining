```python
url_3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
df_3 = pd.read_csv(url_3)
```

```python
def strip_whitespace(df):
    str_cols = df.select_dtypes(include='object').columns
    for col in str_cols:
        df[col] = df[col].str.strip() ### cleaning the whitespace in the columns in all datasets
    return df

cleaned_file1 = strip_whitespace(cleaned_file1)
df_2 = strip_whitespace(df_2)
df_3 = strip_whitespace(df_3)
```

```python
for df in [df_2, df_3]:
    df['customer_lifetime_value'] = df['customer_lifetime_value'].fillna(0).astype(int) ### converting to all the same data type
    df['total_claim_amount'] = df['total_claim_amount'].fillna(0).astype(int)
```
